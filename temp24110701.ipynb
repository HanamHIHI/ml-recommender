```python
import pandas as pd
```

```python
df = pd.read_csv("distance_and_time.csv")
df
```

```python
cond1 = df["total_time"] <= 540
cond2 = df["total_time"] != -1

target_time_df = df.loc[cond1 & cond2]
target_time_df
# print(len(df.loc[cond1 & cond2]))
```

```python
df2 = pd.read_csv("hanam_dict.csv")
df2
```

```python
df3 = pd.merge(left=target_time_df, right=df2, how="inner", left_on='name', right_on='1')[["name", "position", "total_distance", "total_time"]]
df3.to_csv("df_in_10min.csv")
df3
```

```python
df4 = pd.read_csv("category.csv")
df4
```

```python
df5 = pd.merge(df3, df4, how="inner")[["name", "position", "total_distance", "total_time", "category3"]]
df5.to_csv("df_in_10min_v2.csv")
```
